Importing Libraries, dependecies for our program to function:

```python
import numpy as np #we import numpy to work with arrays
import pandas as pd # we import panadas to work and analyse dataframes
from sklearn.model_selection import train_test_split # train_test_split allows us to split our data into training and testing data. Sklearn is a python library from ml data analysis and mining
from sklearn.linear_model import LogisticRegression # we use a logistic regression model hence we import than using the sklearn.linear_model
from sklearn.metrics import accuracy_score # we import accuracy score to evaluate the performance of our learned logistic regression model
```

```python
# loading our csv file and saving the dataset as pandas dataframe. Panada dataframe is a two dimensional tabular structure, labelled rows and columns, which can be removed as wanted
creditcard_data = pd.read_csv('creditcard_2023.csv')
```

```python
#printing the first 5 rows of our dataset, note that the time is the time in seconds from the first trascation taking place
creditcard_data.head() #head funtion gives us the first 5 rows, whereas the tail function gives us the last 5 rows
```

In our table, we see the amount of each trasncation and then its either classified as a fraudualent transcation or a legit transaction, 0 being legit, 1 being fraudualent

```python
#retreiving credit card information
creditcard_data.info()
```

```python
#since there seems to be some missing values, we confirm if there are any missing values or none by using isnull and sum
creditcard_data.isnull().sum()
```

```python
#column 10-28 has some missing values, we have to balance it. We balance it by using imputation which replaces missing values with substiuted ones
from sklearn.impute import SimpleImputer #this imputes values to balance the missing ones
columns_to_imput = creditcard_data.columns[creditcard_data.isnull().any()].tolist() #selects all columns with missing values for imputation
print(columns_to_imput) # to print columns with missing values
#we will use a mean strategy to impute, hence we create an imputer object with mean strategy
imputer = SimpleImputer(strategy='mean')
creditcard_data[columns_to_imput] = imputer.fit_transform(creditcard_data[columns_to_imput]) # we select the missing columns and impute it using our imputer
#now the columns with missing values are imputed and have the imputer as their substitued value
# this block shows the code used to once impute missing values in our data, hwoever we stored the new data without missing values under a cleaned csv file(later in this code ) and hence dont see any missong values now on
#the cleaned data file is stored later on in this code and hence that file has the modified data from the first time we fixed the missing values
```

```python
#we check and verify that no column has missing values
creditcard_data.isnull().sum()
```

```python
#none have a missing value as all show columns show a 0 as non-null values.
#now we check for the distribution of our classified data
creditcard_data['Class'].value_counts() # we check for value count of our class column
```

```python
#we have a present of 0.002229 value for our class in our data, this is an error in our dataset since a transcation can only be legit (0) oe fraudulent (1).
#Since 0.002229 we convert it to 0. We do this by using replace function
creditcard_data['Class'] = creditcard_data['Class'].replace(0.002229,0)
# Ensure that all values in 'Class' are either 0 or 1
creditcard_data['Class'] = creditcard_data['Class'].apply(lambda x: 1 if x == 1 else 0)
#saving cleaned data
creditcard_data.to_csv('creditcard_cleaned.csv', index=False)
```

```python
#we check for classified count again to reconfirm
creditcard_data['Class'].value_counts()
```

```python
#now we balance the data as our two target variables are not balanced. There are 159332 legit transcations and only 356 fraudualent transaction
#We cannot feed this data to our model, hence we balance it
# we seperate the legit and fraudualent dataset for further analysis
legit_data = creditcard_data[creditcard_data.Class == 0]
fraud_data = creditcard_data[creditcard_data.Class == 1]
# we seperate the 0 and 1 column seperately


```

```python
print(legit_data.shape)
print(fraud_data.shape)
```

```python
#statistical data about the transcations, so the amount transcated
legit_data.Amount.describe() #gives transcation information about the legit_data
#count is the amount of legit data, mean is the average of all legit transcations, std is the standard deviation from mean and min is minimum transaction value
#the first, second and third quartiles aswell as maximum legit transaction values is stated
```

```python
fraud_data.Amount.describe() #gives transcation information about the fraud transaction as it did for legit
```

```python
#comparing the two classes of transaction by mean, grouped by mean of every column
creditcard_data.groupby('Class').mean()
```

```python
#the difference in mean of between fraudualent and legit transcation , will help our model identify legit vs fraud transaction
#balancing data, as there are 159332 legit data and 356 fraudualent data, hence we try balancing it
#we use undersampling to balance the data
#we build a sample data set with similar legit and fraudualent transcartion.
# we get 169 legit and fraud data points
legitsample = legit_data.sample(n=356) #we randomly sample 356 data point from 159332
```

```python
# we now join/concantenate the two samples together to make a new sample with equal 0 and 1s
new_dataset = pd.concat([legitsample, fraud_data], axis=0)# axis= 0 makes sure that the fraud_data is added below legit sample, row wise and not column wise
```

```python
# we split the data into test and train set, and train our model with this.
#we check the count of the number of 0 and 1s in our dataset
new_dataset['Class'].value_counts()
```

```python
#checking the new dataset
new_dataset.head()
```

```python
new_dataset.shape #checking the shape of our new dataset
```

```python
new_dataset.groupby('Class').mean()
```

```python
#splitting the dataset into features and targets, to then split them to test and train
```

```python
features = new_dataset.drop(columns='Class', axis=1)
targets = new_dataset['Class']
print(features)
```

```python
print(targets)
```

```python
#splitting features and targets into training data and test data
features_train, features_test, targets_train, targets_test = train_test_split(features, targets, test_size=0.15, stratify=targets, random_state=1)
#we split feaTures into test and train, and target into test and train.
# test data size is 15% of overall data and train data size is 85% of overall data, we stratify to targets so that there are equal numbers of 0 and 1 in both sets, random state ensures that the data is split the same way everytime we run this code
```

```python
print(  features.shape, features_train.shape, features_test.shape)
```

```python
print(targets.shape, targets_train.shape, targets_test.shape)
```

```python
#Training our logistic regression model
LRmodel = LogisticRegression() #defining our model
LRmodel.fit(features_train, targets_train) #training our model, it fits our model to these data

```

```python
#evaluating our model performance
#Accuracy on training data
features_train_prediction = LRmodel.predict(features_train)
training_data_accuracy = accuracy_score(targets_train, features_train_prediction)# compring the real target value comprated to what our model predicts on the training data
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#Our training data accuracy is 95% which is very good
#Accuracy score on test data, evaluating on test data,which is the main part of our program
features_test_prediction = LRmodel.predict(features_test)
test_data_accuracy = accuracy_score(targets_test, features_test_prediction)
print('Accuracy on test data : ', test_data_accuracy)# we compare what our model predicts for test data to what the actual target value for the test data is
```

Our test data accuracy score is 90.65%, which is very good and is not far away from our training data accuracy score. Hence we can say that, our model has not underfitted or overfitted on the training data,  and has good performance and low Generalisation error.
