Importing the following libraries required for our program.


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Numpy is to work with arrays, and pandas(python data analysis) is to work with data processing. The train_test_split function is used to split datasets/matrices/array.We import logisticregression model as thats what we will use for our program, due to the binary classification nature of our model. We import accuracy score to test the accuracy of our models.  We use the sklearn library as it is good for our machine learning algorithm.

Storing data and processing it as a panda dataframe

```python
# loading the data from the file into sonar data using pandas
Sd = pd.read_csv('Copy of sonar data.csv', header = None) # none as we do not have any header on our columns
```

```python
Sd.head()
```

The rows represent an intance of scan, so at one time how the scan was. The column represents the different feature of the scan (at that one instance) so example of features could, be frequency and amplitude. The different rows are different instances of scan for different features and the last column is the label of it either being a rock or the mine (judges each instance differently based on readings of its features)

```python
Sd.shape # gets number of rows and columns
```

```python
# obtaining statistical data about our dataset, gives statistical data for all the columns. Makes us better understand the data.
Sd.describe()
```

```python
Sd[60].value_counts() # tells us how many rock and mine example are there.We check for the 60th column as that is the column containing the rocks and mine
```

The data is well balanced for both rocks and mines as their count is similar and not far apart. M is mine and R is Rocks

```python
#Grouping the similar types together. This will give us the difference in mean of Mine and Rock for each column
Sd.groupby(60).mean()
```

Spliting label from data to store label in a different variable

```python
#seperating label and features
X_NonLabel = Sd.drop(columns= 60)
X_Label = Sd[60]
print(X_NonLabel)
print(X_Label)
```

We now split noth sets of data X_NonLabel and X_Label into training and test set.

```python
# we split the data into train and test and gave those data their label accordingly, for both train and test data. Test size assigns the size of the test data as a percentage of the whole dataset, in this case 10% of data is test data.
# We use stratify to have equal number of rocks and mine in both test and training data
#Random state ensures that the data is split the same way everytime we run it, this helps with debugging and model performance. To ensure same split each time, we use random state.
X_NonLabel_Train,X_NonLabel_Test, X_Label_Train, X_Label_Test = train_test_split(X_NonLabel, X_Label, test_size = 0.1, stratify = X_Label, random_state = 1)
```

```python
#checking the amount of data in our train and test set
print(X_NonLabel.shape,X_NonLabel_Train.shape, X_NonLabel_Test.shape)
# we have 187 data to train and 21 data to test
```

Now, we finally move to the logistic regression training of the model. We already imported the logistic regression model, we will train it with our training data.

```python
logistic_model = LogisticRegression()
# initialising the model
logistic_model.fit(X_NonLabel_Train, X_Label_Train)
#We train the model with X_NonLabel_Train, X_Label_Train
```

```python
#Model Evaluation for training data, accuracy check. The model learnt on the training data, we will see how well it predicts that dataset
#Generally the accuracy is higher with more data points
X_NonLabel_Train_prediction = logistic_model.predict(X_NonLabel_Train) #we predict the label of our training data
training_data_accuracy = accuracy_score(X_Label_Train, X_NonLabel_Train_prediction) #we compare our model prediction with real label
```

```python
print("The training data accuracy score is:",training_data_accuracy )
```

Therefore, our training data accuracy score is 0.8342 which is 83.42%. This is pretty high and a good accuracy score, which is expected as our model learnt on the training data.

```python
#Accuracy score on our test data, how accurate is our test data
X_NonLabel_Test_prediction = logistic_model.predict(X_NonLabel_Test)
#Prediction of our test data by our logistic regression model
test_data_accuracy = accuracy_score(X_Label_Test, X_NonLabel_Test_prediction)
#comparing our prediction using our logistic model compared to the real label of that test data
```

```python
print("The test data accuracy score is:", test_data_accuracy)
```

We have an test accuracy of 76.2%, which is very good.

Constructing a predictive system, using our logistic model to predict if its a rock or mine.

```python
# Getting input from user, then splitting it by spaces, converting into float and hten to a numpy array using np
user_values = input("Enter the values seperated by either a space or a comma")
# we replace the commas with spaces and split the string by spaces
values_str = user_values.replace(',',' ').split()
# converting strings to float
values = [float(value) for value in values_str]
# we convert it into numpy array to process better
#converting input data list to numpy array
input_numpy = np.asarray(values)
```

```python
# since we take input for one instance, we will reshape it for one instance. We use one instance to predict for that one instance
input_reshape = input_numpy.reshape(1,-1)
# 1, -1 represents one instance, we predict for that one instance
```

```python
# prediction for the instance
prediction = logistic_model.predict(input_reshape)
# our model is labelled as logistic_model and our data is input_reshape so we use them to predict for the data, input_reshape
print(prediction)
```
